```python
import pandas as pd

df = pd.read_csv("Customer_data.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df['gender'] = df['gender'].map({'Male': 1, 'Female': 0})
```

```python
df = df.drop_duplicates()
```

```python
for col in df.columns:
    print(col, df[col].unique())
```

```python
df = df.rename(columns={
    'no_of_orders_placed': 'demand',
    'restaurnat_rating': 'restaurant_rating'
})
```

```python
# Remove wrong value
df = df[df['delivery_time'] != 'Delivery Time']

# Convert to numeric
df['delivery_time'] = df['delivery_time'].astype(int)
```

```python
mapping = {
    'Strongly disagree': 1,
    'Disagree': 2,
    'Neutral': 3,
    'Agree': 4,
    'Strongly agree': 5
}

ordinal_cols = [
    'ease_and_convenient',
    'self_cooking',
    'health_concern',
    'late_delivery',
    'poor_hygiene',
    'bad_past_experience',
    'more_offers_and_discount'
]

for col in ordinal_cols:
    df[col] = df[col].map(mapping)
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.info()
df.head()
```

```python
cols_to_drop = [
    'ease_and_convenient',
    'self_cooking',
    'health_concern',
    'late_delivery',
    'poor_hygiene',
    'bad_past_experience',
    'more_offers_and_discount'
]

df = df.drop(columns=cols_to_drop)
```

```python
df = df.rename(columns={
    'no._of_orders_placed': 'demand'
})
```

```python
df.info()
```

```python
df = df.reset_index(drop=True)
```

```python
df.shape
df.info()
```

```python
df.to_csv("cleaned_food_delivery_data.csv", index=False)
```

## 2. Exploratory Data Analysis (EDA) & Descriptive Statistics

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure()
df['demand'].hist()
plt.title("Distribution of Order Demand")
plt.xlabel("Demand")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='delivery_time', y='demand', data=df)
plt.title("Demand vs Delivery Time")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='restaurant_rating', y='demand', data=df)
plt.title("Demand by Restaurant Rating")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='delivery_rating', y='demand', data=df)
plt.title("Demand by Delivery Rating")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
df['demand'].hist()
plt.title("Distribution of Demand")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='gender', y='demand', data=df)
plt.title("Demand by Gender")
plt.show()
```

```python
df.groupby('restaurant_rating')['demand'].mean().plot(kind='bar')
plt.title("Average Demand by Restaurant Rating")
plt.ylabel("Average Demand")
plt.show()
```
